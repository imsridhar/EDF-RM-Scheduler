```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(20,10))
# x axis values 
rm = [30,62,79,88,94,95,104,102,101,97,93,90,86,83,78]
# corresponding y axis values 
x = [i*4 for i in range(1,16)]  
edf = [0,1,6,14,18,25,40,52,62,74,80,78,77,73,68]  
# plotting the points  
plt.plot(x, rm, color='green', linestyle='dashed', linewidth = 3, 
         marker='o', markerfacecolor='blue', markersize=12, label = "RM") 
plt.plot(x, edf, color='red', linestyle='dashed', linewidth = 3, 
         marker='*', markerfacecolor='purple', markersize=12,label = "EDF")
# naming the x axis 
plt.xlabel('Number of tasks') 
# naming the y axis 
plt.ylabel('Average number of preemptions') 
  
  
# function to show the plot 
plt.show()
```
